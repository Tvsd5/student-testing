# Задание 1

Найдите сумму чисел **2 и 3**.

Введите ответ ниже и запустите ячейки по порядку.

```python
answer = input("Введите ответ: ")
```

```python
import js
import json

async def submit(answer):
    response = await js.fetch(
        "https://student-checker.tiyfre.workers.dev",
        {
            "method": "POST",
            "headers": {"Content-Type": "application/json"},
            "body": json.dumps({"answer": answer})
        }
    )

    result = await response.json()

    if result.get("correct"):
        print("✅ Правильно!")
    else:
        print("❌ Неправильно")

submit(answer)
```

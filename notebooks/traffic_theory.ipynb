```
student_id = input("Введите ваш ID (номер зачётки или логин): ")
```

```
import json
from pyodide.http import pyfetch

WORKER_URL = "https://student-checker.tiyfre.workers.dev"  # Замените на свой URL

async def submit(task_id, answer):
    response = await pyfetch(
        WORKER_URL,
        method="POST",
        body=json.dumps({
            "student_id": student_id,
            "task_id": task_id,
            "answer": answer
        })
    )
    return await response.json()
```

## Задание 1. Интенсивность входного потока

Исходные данные задаются сервером. Решите задачу и введите ответ.

```
answer = input("Введите ответ: ")
```

```
result = await submit(1, answer)
print(result["message"])
if not result["correct"]:
    print("Новые исходные данные:", result.get("params"))
```

‚ö†Ô∏è **–ò–Ω—Å—Ç—Ä—É–∫—Ü–∏—è —Å—Ç—É–¥–µ–Ω—Ç—É**

- –î–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–¥–∞–Ω–∏—è –Ω–∞–∂–º–∏—Ç–µ ‚ñ∂Ô∏è **Run** —É –Ω—É–∂–Ω–æ–≥–æ –∑–∞–¥–∞–Ω–∏—è
- –í–≤–µ–¥–∏—Ç–µ –æ—Ç–≤–µ—Ç –∏ –Ω–∞–∂–º–∏—Ç–µ Enter
- –ü—Ä–∏ –æ—à–∏–±–∫–µ –±—É–¥–µ—Ç –≤—ã–¥–∞–Ω –Ω–æ–≤—ã–π –≤–∞—Ä–∏–∞–Ω—Ç
- –ö–æ–¥ –∏–∑–º–µ–Ω—è—Ç—å **–Ω–µ –Ω—É–∂–Ω–æ**


```
student_id = input("–í–≤–µ–¥–∏—Ç–µ –≤–∞—à –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä: ")
```

```
# –°–õ–£–ñ–ï–ë–ù–´–ô –ö–û–î (–ù–ï –ò–ó–ú–ï–ù–Ø–¢–¨)
import json
from pyodide.http import pyfetch

WORKER_URL = "https://student-checker.tiyfre.workers.dev"

async def run_task(task_id):
    while True:
        r = await pyfetch(
            WORKER_URL + "/get_task",
            method="POST",
            headers={"Content-Type": "application/json"},
            body=json.dumps({
                "student_id": student_id,
                "task_id": task_id
            })
        )
        task = await r.json()

        print("\n" + "="*60)
        print(f"–ó–ê–î–ê–ù–ò–ï {task_id}")
        print(task["text"])
        print("–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
        for k, v in task["params"].items():
            print(f"  {k} = {v}")
        print("="*60)

        answer = float(input("–í–≤–µ–¥–∏—Ç–µ –æ—Ç–≤–µ—Ç: "))

        r = await pyfetch(
            WORKER_URL + "/submit_answer",
            method="POST",
            headers={"Content-Type": "application/json"},
            body=json.dumps({
                "student_id": student_id,
                "task_id": task_id,
                "answer": answer
            })
        )
        result = await r.json()
        print(result["message"])

        if result["correct"]:
            break
```

# Теория телетрафика — контрольные задания

В каждом задании:
- сервер выдаёт исходные данные
- вы рассчитываете ответ
- при ошибке выдаётся новый вариант

Работайте последовательно.

```
student_id = input("Введите ваш ID (зачётка или логин): ")
```

```
import json
from pyodide.http import pyfetch

WORKER_URL = "https://student-checker.tiyfre.workers.dev"

async def get_task(task_id):
    r = await pyfetch(
        WORKER_URL + "/get_task",
        method="POST",
        body=json.dumps({"student_id": student_id, "task_id": task_id})
    )
    return await r.json()

async def submit_answer(task_id, answer):
    r = await pyfetch(
        WORKER_URL + "/submit_answer",
        method="POST",
        body=json.dumps({
            "student_id": student_id,
            "task_id": task_id,
            "answer": answer
        })
    )
    return await r.json()
```

## Задание 1. Интенсивность входного потока

```
task = await get_task(1)
print(task["text"])
print("Исходные данные:")
for k,v in task["params"].items():
    print(f"  {k} = {v}")
```

```
answer = float(input("Введите ответ: "))
```

```
result = await submit_answer(1, answer)
print(result["message"])
if not result["correct"]:
    print("Попробуйте ещё раз — будет новый вариант.")
```
